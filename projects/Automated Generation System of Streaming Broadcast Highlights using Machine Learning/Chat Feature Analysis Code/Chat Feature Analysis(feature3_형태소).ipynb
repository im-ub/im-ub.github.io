```python
konall1 = pd.merge(kon_sum1,kon_HL, how='left',on = '형태소')
konall1
```

```python
konall = pd.merge(konall1,kon_NHL, how='left',on = '형태소')
konall
```

```python
konall = konall.fillna(0)
```

```python
word = konall['형태소']
word
```

```python
def shorthead(sentence):
    x = len(sentence)
    for i in range(1, (x // 2)+1):
        if(x % i == 0):
            s = sentence[0:i]
            count = int(x / i)
            if (s*count == sentence):
                return s + s
    return sentence
```

```python
shorthead(('ㅊㅊㅊㅊㅊㅊㅊㅊ', 'KoreanParticle'))
```

```python
word2 = []
for i in word:
    if 'KoreanParticle' in i:
        t = i.split('/')[0]
        h = shorthead(t)
        s = h + '/KoreanParticle'
        word2.append(s)
    else:
        word2.append(i)
```

```python
len(word2)
```

```python
dfword = pd.DataFrame(word2)
```

```python
dfword.rename(columns={ 0:'형태소2'}, inplace = True)
dfword
```

```python
konword = pd.merge(dfword, konall, left_index = True, right_index = True, how = 'left')
konword
```

```python
groupsum = konword.groupby('형태소2').sum()
groupsum
```

```python
wordsum = pd.merge(dfword, groupsum, how = 'left', on = '형태소2')
wordsum
```

```python
for i,row in wordsum.iterrows():
    if wordsum.at[i,'하이라이트 형태소수'] == 0:
        wordsum.at[i,'하이라이트 비율'] = 0
    else:
        wordsum.at[i,'하이라이트 비율'] = wordsum.at[i,'하이라이트 형태소수'] / wordsum.at[i,'전체 형태소수']
wordsum
```

```python
for i,row in wordsum.iterrows():
    if wordsum.at[i,'비하이라이트 형태소수'] == 0:
        wordsum.at[i,'비하이라이트 비율'] = 0
    else:
        wordsum.at[i,'비하이라이트 비율'] = wordsum.at[i,'비하이라이트 형태소수'] / wordsum.at[i,'전체 형태소수']
wordsum
```

```python
for i,row in wordsum.iterrows():
    if wordsum.at[i,'하이라이트 비율'] <= wordsum.at[i,'비하이라이트 비율']:
        wordsum.at[i,'Hscore'] = 0
    else:
        wordsum.at[i,'Hscore'] = 1 - (wordsum.at[i,'비하이라이트 비율'] / wordsum.at[i,'하이라이트 비율'])
wordsum
```

```python
for i,row in wordsum.iterrows():
    if wordsum.at[i,'비하이라이트 비율'] <= wordsum.at[i,'하이라이트 비율']:
        wordsum.at[i,'Nscore'] = 0
    else:
        wordsum.at[i,'Nscore'] = 1 - (wordsum.at[i,'하이라이트 비율'] / wordsum.at[i,'비하이라이트 비율'])
wordsum
```

```python
total_score = wordsum.loc[:,["Hscore","Nscore"]]
total_score
```

```python
lal = pd.read_csv("형태소_풍월량_926484159.csv", encoding = 'cp949')
lal
```

```python
lal = lal.loc[:,["형태소2","Hscore","Nscore"]]
lal
```

```python
score = pd.merge(rat, lal, how ='outer', on='형태소2')
score
```

```python
score = score.fillna(0)
score
```

```python
score.columns
```

```python
for i,row in score.iterrows():
    score.at[i,'Hscore'] = (score.at[i,'Hscore_x'] +score.at[i,'Hscore_y'])/2
    score.at[i,'Nscore'] = (score.at[i,'Nscore_x'] +score.at[i,'Nscore_y'])/2
score    
```

```python
score_board = score.drop(['Hscore_x','Hscore_y','Nscore_x','Nscore_y'],axis = 1)
score_board
```

```python
score_board.to_csv("Scoreboard.csv",encoding ='cp949')
```

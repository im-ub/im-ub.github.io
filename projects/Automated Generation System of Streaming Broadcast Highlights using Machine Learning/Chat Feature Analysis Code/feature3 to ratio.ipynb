```python
data = pd.read_csv("[Final Window]풍월량_926484159.csv", encoding='cp949' )
data
```

```python
fill_val1 = {'채팅내용':"",
           '시청자ID':"",
           '하이라이트 여부':"X"}
```

```python
data=data.fillna(fill_val1)
data
```

```python
df1 = data["채팅내용"]
df1
```

```python
df1 = data["채팅내용"]
chat1 = df1.to_numpy()
chat1
```

```python
arr = []
for h in chat1:
    pos_chat = okt.pos(h, join=True)
    arr.append(pos_chat)
```

```python
arr
```

```python
ar = []
for s in arr:
    count = Counter(s)
    ar.append(count)
```

```python
ar
```

```python
kon = pd.DataFrame(ar)
```

```python
kon
```

```python
kon=kon.fillna(0)
```

```python
ws = kon.sum(axis=1)
ws
```

```python
wordsum = pd.DataFrame(ws)
wordsum
```

```python
wordsum.rename(columns= { 0:'채팅당 형태소수'}, inplace=True)
wordsum
```

```python
# data2 = pd.read_csv("[Final Window]괴물쥐_935176016.csv", encoding='cp949')
# data2
```

```python
window = pd.merge(data, wordsum, left_index=True, right_index=True, how='left' )
window
```

```python
for i,row in window.iterrows():
    if window.at[i,'채팅당 형태소수'] == 0:
#         window.at[i,'Hscore'] = window.at[i,'Hscore']
#         continue
        window.at[i,'Hscore'] = 0
        window.at[i,'Nscore'] = 0
    else:
        window.at[i,'Hscore'] = window.at[i,'Hscore']/window.at[i,'채팅당 형태소수']
        window.at[i,'Nscore'] = window.at[i,'Nscore']/window.at[i,'채팅당 형태소수']
```

```python
window
```

```python
final_window = window.drop('채팅당 형태소수',axis=1)
final_window
```

```python
final_window.to_csv("[Final Window]풍월량_926484159.csv", encoding='cp949')
```

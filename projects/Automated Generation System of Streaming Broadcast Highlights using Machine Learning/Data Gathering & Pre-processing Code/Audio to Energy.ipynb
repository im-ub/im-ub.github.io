## Audio to Energy (using librosa: onset_strength) 
### Audio data pre-processing

```python
%matplotlib inline
import matplotlib.pyplot as plt, IPython.display as ipd, numpy
import librosa, librosa.display
# import stanford_mir; stanford_mir.init()

x, sr = librosa.load('C:\\Users\\ub16\\Documents\\notebook\\audio_convert\\20210214_913699132.wav')

print(x.shape, sr)
```

```python
hop_length = 22050
onset_envelope = librosa.onset.onset_strength(x, sr=sr, hop_length=hop_length)
```

```python
onset_envelope.shape
```

```python
N = len(x)
T = N/float(sr)
t = numpy.linspace(0, T, len(onset_envelope))
```

```python
plt.figure(figsize=(14, 5))
plt.plot(t, onset_envelope)
plt.xlabel('Time (sec)')
plt.xlim(xmin=0)
plt.ylim(0)
```

```python
onset_envelope.to_csv("20210214_913699132.csv")
```
